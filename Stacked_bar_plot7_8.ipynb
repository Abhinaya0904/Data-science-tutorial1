```python
import matplotlib.pyplot as plt
#Defining categories and values for two gropus
categories = ['Transport','Income Tax','Housing']
values1 = [15, 24, 30]
values2 = [20, 18, 25]
#creating the first set of bars (Group 1) without any offset
plt.bar(categories,values1, label='First Half', color='skyblue')
#creating the second set of bars (Group2) plotted with
#This makes Group 2 bars stacked on the top of Group 1 bars
plt.bar(categories,values2, bottom=values1, label='Second Half', color='Yellow')
#Adding labels to the axes
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Graph')
plt.legend()
plt.show()
```

```python
#Histogram

import matplotlib.pyplot as plt
x = [70, 50, 100, 70, 50, 100, 50, 50, 50, 70, 50]
plt.hist(x, orientation="vertical")
plt.title('Mark Analysis')
plt.ylabel('No of Students')
plt.xlabel('Marks')
plt.show()
```

```python
#cummulative histogram

import matplotlib.pyplot as plt
import numpy as np
#Generate random exam scores (out of 100)
exam_scores = np.random.randint(0, 100, 150)
#Create a cummulative Histogram
plt.hist(exam_scores, bins=20, cumulative=True, color='Yellow', edgecolor='black')
plt.xlabel('Exam scores')
plt.ylabel('Cumulative no of students')
plt.title('Cumulative Histogram of Exam Scores')
plt.show()
```

```python
#STACKED HISTOGRAM

import matplotlib.pyplot as plt
import numpy as np
#Sample data for two data sets
mark1 = np.array([2,4,5,7,9,10,11,13,14,15])
mark2 = np.array([6,7,8,10,11,12,13,14,15,16])
#Creating a stacked histogram with different colors
plt.hist([mark1,mark2],bins=20,stacked=True,color=['skyblue','salmon'],edgecolor='black')
plt.xlabel('marks')
plt.ylabel('Count')
plt.title('Stacked Histogram with Multiple Datasets')
plt.legend(['Mark Dataset 1','Mark Dataset 2'])
plt.show()
```

```python
#PANDAS

import pandas as pd
#Series generated from a list
a = [1, 7, 2]
myvar = pd.Series(a)
print(a)
print(myvar)
print('-----------------------------------')
#LABELS
print(myvar[0])
print('-----------------------------------')
#Assinging seperate labels to values in the
myvar2 = pd.Series(a, index = ["x", "y", "z"])
print(myvar2)
print(myvar2["y"])
print('-----------------------------------')
#Series generated from dictoniary
calories = {"day1": 420, "day2": 380, "day3": 390}
myvar = pd.Series(calories)
print(myvar)
print('------------------------------------')
myvar = pd.Series(calories, index =["day1","day2"])
print(myvar)
print('------------------------------------')
#series generated from scalar value
series = pd.Series(5, index=[0, 1, 2, 3])
print(series)
print('---------------------------------')

```

```python
import pandas as pd
#Integer Series
int_series = pd.Series([1, 2, 3, 4])
print(int_series)
print('----------------------------')
#float Series
float_series = pd.Series([1.1, 2.2, 3.3, 4.4])
print(float_series)
print('----------------------------')
#string Series
string_series = pd.Series(['a', 'b', 'c', 'd'])
print(string_series)
print('-----------------------------')
#Label-based indexing
data = {'a':10, 'b': 20, 'c': 30}
series = pd.Series(data)
print(series['a'])
print(series[['a', 'c']])
print('-----------------------------')
#Boolean indexing
data = [10, 20, 30, 40, 50]
series = pd.Series(data)
#Access elements greater than 25
print(series[series > 25])
print('------------------------------')
```
